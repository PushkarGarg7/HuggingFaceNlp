```python
from transformers import pipeline
```

```python
classifier=pipeline("feature-extraction")
classifier("I like football and dislike cricket.")
```

```python
classifier=pipeline("zero-shot-classification")
classifier("I am going to spain to watch a match.",
           candidate_labels=["football","basketball","cricket"])
```

```python
classifier=pipeline("question-answering")
classifier(
    question="what is my name?",
    context="they call me Pushkar."
)['answer']
```
